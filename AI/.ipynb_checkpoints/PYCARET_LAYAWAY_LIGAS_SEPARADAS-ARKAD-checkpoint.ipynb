```python
import pandas as pd
from pycaret.classification import *
import numpy as np
from openpyxl import Workbook, load_workbook
from xlsxwriter import Workbook
from time import sleep
from datetime import date, datetime, timedelta

def drop_reset_index(df):
    df = df.dropna()
    df = df.reset_index(drop=True)
    df.index += 1
    return df
```

```python
caminho1 = "C:/Users/thiag/OneDrive/curso python/modelo_leandro_2/"
jogos_dia_completo = pd.read_csv(caminho1 + 'jogos_estatistica.csv')

dia = date.today()
# + timedelta(1)

jogos_do_dia = pd.read_excel("C:/Users/thiag/OneDrive/curso python/modelo_leandro_2/2023-07-04_Jogos_do_Dia_arkad_transform(last).xlsx")
```

```python
flt = jogos_do_dia.League == "USA - MLS"
jogos_do_dia = jogos_do_dia[flt]
jogos_do_dia = drop_reset_index(jogos_do_dia)
saved_final_lr = load_model('LAYAWAY_ARKAD_USA - MLS')
new_predictions = predict_model(saved_final_lr, data=jogos_do_dia)
new_predictions
```

```python
base_H = new_predictions[['p_H', 'CV_VG_H', 'cv_pontos']]
    
base_A = dnew_predictions[['p_A', 'CV_VG_A', 'cv_pontos']]    
```

```python
lista=[]

for a,b,c,d,e,f,g,h,i,j,k,l in zip(jogos_do_dia.League,jogos_do_dia.Date,jogos_do_dia.Time,jogos_do_dia.Home,jogos_do_dia.Away,
                                   jogos_do_dia.FT_Odd_H,jogos_do_dia.FT_Odd_D,jogos_do_dia.FT_Odd_A,jogos_do_dia.FT_Odd_Over25,jogos_do_dia.FT_Odd_Under25,
                                   jogos_do_dia.FT_Odd_BTTS_Yes,jogos_do_dia.FT_Odd_BTTS_No):
        League = a
        Date = b
        Time = c 
        home = d
        away = e
        FT_Odd_H = f
        FT_Odd_D = g
        FT_Odd_A = h
        FT_Odd_Over25 = i
        FT_Odd_Under25 = j
        FT_Odd_BTTS_Yes = k
        FT_Odd_BTTS_No = l
  
        df1 = base_H[base_H.cv_pontos == home].tail(1)

        df2 = base_A[base_A.cv_pontos == away].tail(1)

        jogo = {}

        jogo["League"] = League 
        jogo["Date"] = Date
        jogo["Time"] = Time
        jogo["Home"] = home
        jogo["Away"] = away
        jogo["FT_Odd_H"] = FT_Odd_H
        jogo["FT_Odd_D"] = FT_Odd_D
        jogo["FT_Odd_A"] = FT_Odd_A
        jogo["FT_Odd_Over25"] = FT_Odd_Over25
        jogo["FT_Odd_Under25"] = FT_Odd_Under25
        jogo["FT_Odd_BTTS_Yes"] = FT_Odd_BTTS_Yes
        jogo["FT_Odd_BTTS_No"] = FT_Odd_BTTS_No

        
        try:
        
            jogo['p_H'] = df1[df1.cv_pontos == home]['p_H'].iloc[0]
            jogo['CV_VG_H'] = df1[df1.cv_pontos == home]['CV_VG_H'].iloc[0]
        
            jogo['CV_VG_A'] = df2[df2.Away == away]['CV_VG_A'].iloc[0]
            jogo['p_A'] = df2[df2.Away == away]['p_A'].iloc[0]
    
            
                
            
        except:
            pass
        
        lista.append(jogo)


df = pd.DataFrame(lista)
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)


```
