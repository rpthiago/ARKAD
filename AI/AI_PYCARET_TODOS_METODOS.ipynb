```python
import pandas as pd
from pycaret.classification import *
import numpy as np
from openpyxl import Workbook, load_workbook
from xlsxwriter import Workbook
```

```python
caminho1 = "C:/Users/thiag/OneDrive/curso python/modelo_leandro_2/"

df_jogos = pd.read_csv(caminho1 + 'jogos_estatistica.csv')
```

```python
#LAYDRAW

saved_final_lr = load_model('LAYDRAW_LR_TODOS_JOGOS')

laydraw = predict_model(saved_final_lr, data = df_jogos)

df4 = pd.merge(df_jogos, laydraw, how='left', on=["id"])
df4 = df4[['id', 'date_unix', 'Liga', 'home_name', 'away_name', 'odds_ft_1', 'odds_ft_x', 'odds_ft_2', 'prediction_label']]

laydraw = ((df4['prediction_label'] == 1))

df4_laydraw = df4[laydraw]     
```
